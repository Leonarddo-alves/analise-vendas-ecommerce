```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)

df.head()
```

```python
df["Order Month"] = df["Order Date"].dt.to_period("M")
df["Order Year"] = df["Order Date"].dt.year
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days

df[["Order Date", "Order Month", "Order Year", "Shipping Days"]].head()
```

```python
total_sales = df["Sales"].sum()
print(f"Faturamento total: USD {total_sales:,.2f}")
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
sales_by_category
```

```python
sales_by_category.plot(kind="bar")

plt.title("Vendas por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Total de Vendas")
plt.xticks(rotation=45)
plt.show()
```

A categoria Technology apresentou o maior faturamento, com aproximadamente USD 827 mil em vendas. Em seguida aparecem Furniture e Office Supplies, com valores relativamente próximos.


```python
sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
sales_by_region
```

```python
sales_by_region.plot(kind="bar")

plt.title("Vendas por Região")
plt.xlabel("Região")
plt.ylabel("Total de Vendas")
plt.xticks(rotation=45)
plt.show()
```

A região West apresentou o maior faturamento, com aproximadamente USD 710 mil em vendas. A região South teve o menor faturamento, com cerca de USD 389 mil, indicando uma possível oportunidade de investigação sobre desempenho regional.

```python
sales_by_month = df.groupby("Order Month")["Sales"].sum()
sales_by_month
```

```python
sales_by_month.plot(kind="line", marker="o")

plt.title("Vendas por Mês")
plt.xlabel("Mês")
plt.ylabel("Total de Vendas")
plt.xticks(rotation=45)
plt.show()
```

2018-11: USD 117,938.16

As vendas apresentam crescimento em alguns períodos do segundo semestre, especialmente entre setembro e dezembro. O maior faturamento mensal ocorreu em novembro de 2018, com aproximadamente USD 117,9 mil em vendas.

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_products.sort_values().plot(kind="barh")

plt.title("Top 10 Produtos por Vendas")
plt.xlabel("Total de Vendas")
plt.ylabel("Produto")
plt.show()
```

O produto com maior faturamento foi o Canon imageCLASS 2200 Advanced Copier, com aproximadamente USD 61,6 mil em vendas. Ele teve um desempenho bem acima dos demais produtos do top 10, indicando que produtos de tecnologia de maior valor podem ter grande impacto no faturamento total.


```python
sales_by_segment = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)
sales_by_segment
```

```python
sales_by_segment.plot(kind="bar")

plt.title("Vendas por Segmento")
plt.xlabel("Segmento")
plt.ylabel("Total de Vendas")
plt.xticks(rotation=45)
plt.show()
```

O segmento Consumer foi responsável pelo maior faturamento, com aproximadamente USD 1,15 milhão em vendas. Em seguida aparecem Corporate, com cerca de USD 688 mil, e Home Office, com aproximadamente USD 425 mil.


Conclusão

A análise mostrou que o dataset possui um faturamento total de aproximadamente USD 2,26 milhões. A categoria Technology foi a que mais contribuiu para as vendas, seguida por Furniture e Office Supplies.

Em relação às regiões, West apresentou o maior faturamento, enquanto South teve o menor desempenho. Também foi possível observar que as vendas tendem a crescer em alguns meses do segundo semestre, principalmente entre setembro e dezembro.

O produto com maior faturamento foi o Canon imageCLASS 2200 Advanced Copier, com vendas muito acima dos demais produtos. Já no perfil dos clientes, o segmento Consumer foi o principal responsável pelo faturamento total.

Esses resultados indicam que produtos de tecnologia, clientes do segmento Consumer e a região West são pontos importantes para o desempenho comercial da empresa.
